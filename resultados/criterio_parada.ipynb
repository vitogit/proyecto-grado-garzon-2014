```python
import re
#%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = []
y = np.arange(501)

files = []
files.append(open('garzon20_32.o9290', 'r'))
files.append(open('garzon48.o9421', 'r'))
files.append(open('garzon64.o9413', 'r'))

addValue = True

#filew = open("32.txt", "w")
for file in files:
    for line in file:
        m = re.search('Best cost trial: (\d*\.\d*)', line)
        ct = re.search('Current trial: (\d*)', line)
        if ct:
            ctValue = ct.group(1)
            if ctValue == "1":
                addValue = False
            else:
                addValue = True
                
        if m and addValue:
            found = m.group(1)
            x.append(found)
            addValue = True
            #filew.write(found+"\n")

        if len(x) == 501:
            plt.plot(y, x)
            x=[]

    file.close()

#filew.close()

#http://stackoverflow.com/questions/4805048/how-to-get-different-colored-lines-for-different-plots-in-a-single-figure
plt.xlabel('Generaciones')
plt.ylabel('Fitness')
plt.title('Criterio de parada : 500 generaciones')


#plt.legend(['y = x', 'y = 2x'], loc='upper left')
plt.savefig('criterio_parada.png')
plt.show()
```
